```python
#list all files in /Users/carbs/mkb102/icd-10_2016_to_icd-10-am_eleventh_edition (1)

import os

# Get all files in the directory
# Get directory path
dir_path = '/Users/carbs/mkb102/icd-10_2016_to_icd-10-am_eleventh_edition'

# List files in directory
files = os.listdir(dir_path)

# #"Counter","ICD-10 code","WHO update 1=Add 2=Del","ICD-10 code descriptor","ICD-10-AM Map","Aust update 1=Add 2=Del","Additive Map (+)","ICD-10-AM code descriptor"
# 1,"A00.0",,"Cholera due to Vibrio cholerae 01, biovar cholerae","A00.0",,,"Cholera due to Vibrio cholerae 01, biovar cholerae"
# 2,"A00.1",,"Cholera due to Vibrio cholerae 01, biovar eltor","A00.1",,,"Cholera due to Vibrio cholerae 01, biovar eltor"


all_dfs = []
for file in files:
    # Get absolute path by joining directory path with filename
    file_path = os.path.join(dir_path, file)
    # Read txt file
    df = pd.read_csv(file_path, encoding="latin1", dtype=str, keep_default_na=False,)
    print(df.columns)
    #PRINT FIRST LINE OF THE FILE
    print(df.iloc[0])
    all_dfs.append(df)

#create a dataframe from all the dataframes
all_mappings_df = pd.concat(all_dfs)

#save the dataframe to a csv fil

#strip the ICD-10-AM Map column
all_mappings_df['ICD-10-AM Map'] = all_mappings_df['ICD-10-AM Map'].str.strip()





```

```python
all_mappings_df['ICD-10-AM Map'].isna().sum()

#drop the rows where the 'ICD-10-AM Map' column is na
all_mappings_df = all_mappings_df[all_mappings_df['ICD-10-AM Map'].notna()]
```

```python
all_mappings_df['ICD-10-AM Map'].value_counts()
```

```python
#remove duplicate ICD-10-AM Map
all_mappings_df = all_mappings_df.drop_duplicates(subset=['ICD-10-AM Map'])




```

```python
all_mappings_df['ICD-10-AM Map'].value_counts()
```

```python
import re
import json

mkb_slo_df_eng = mkb_slo_df.copy()
mkb_slo_df_eng['ENGLISH DESCRIPTION'] = ''

# Get length distribution of ICD-10-AM Map codes
test = all_mappings_df['ICD-10-AM Map'].apply(lambda x: len(x))
print("ICD-10-AM Map code lengths:")
print(test.value_counts().sort_index())
print("\nOutlier codes (lengths > 5 or < 3):")
print(all_mappings_df[test > 5]['ICD-10-AM Map'].tolist())

# Get length distribution of Slovenian codes
test = mkb_slo_df_eng['KODA'].apply(lambda x: len(x))
print("\nSlovenian code lengths:")
print(test.value_counts().sort_index())
print("\nOutlier codes (lengths > 5 or < 3):")
print(mkb_slo_df_eng[test > 5]['KODA'].tolist())

# Create mapping dictionary for faster lookup
icd_map = dict(zip(all_mappings_df['ICD-10-AM Map'], all_mappings_df['ICD-10-AM code descriptor']))

# Vectorized update of English descriptions
mkb_slo_df_eng['ENGLISH DESCRIPTION'] = mkb_slo_df_eng['KODA'].map(icd_map)
```

```python
#apply another zero and do it again
mkb_slo_df_eng['KODA_STANDARDIZED_2'] = mkb_slo_df_eng['KODA_STANDARDIZED'].apply(lambda x: x+"0")

mkb_slo_df_eng['ENGLISH DESCRIPTION_FROM_STANDARDIZED_2'] = mkb_slo_df_eng['KODA_STANDARDIZED_2'].map(icd_map)

#combine english description and english description from standardized
mkb_slo_df_eng['ENGLISH DESCRIPTION_COMBINED_2'] = mkb_slo_df_eng['ENGLISH DESCRIPTION'].fillna(mkb_slo_df_eng['ENGLISH DESCRIPTION_FROM_STANDARDIZED']).fillna(mkb_slo_df_eng['ENGLISH DESCRIPTION_FROM_STANDARDIZED_2'])



#check if we have missing values
print(len(mkb_slo_df_eng[mkb_slo_df_eng['ENGLISH DESCRIPTION_COMBINED_2'].isna()]['KODA'].tolist()))

#check if we have missing values
print(mkb_slo_df_eng[mkb_slo_df_eng['ENGLISH DESCRIPTION_COMBINED_2'].isna()]['KODA'].tolist())

```
